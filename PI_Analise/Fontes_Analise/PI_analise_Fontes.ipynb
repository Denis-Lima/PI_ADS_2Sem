# Analise dos dados fornecidos pelo spc brasil
## Filtro de erros

```python
##importando pandas
import pandas as pd
import requests as rq
```

```python
##abrir o arquivo STG_FNT_ITT Conceito: Extração das informações da tabela “fonte_cad_positivo”.
##Cadastro de todas as fontes que disponibilizam dados para consumo do cadastro positivo.
##Atualmente sua maioria é composta por bancos.
##Campos disponíveis no arquivo: ID_STG_FNT_ITT, NUM_CNPJ, NUM_CMP_CNPJ, NOM_COM, NOM_RAZ_SCL e DAT_INC_DBO
df_fontes = pd.read_excel("STG_FNT_ITT.xlsx")
df_fontes.head(10)
```

```python
### Ja sei que alguns itens do cnpj estão diferentes de 14 caracteres
df_fontes.dtypes
```

```python
###converter cnpj para string
df_fontes['NUM_CNPJ'] = df_fontes.NUM_CNPJ.astype('str')
df_fontes['NUM_CMP_CNPJ'] = df_fontes.NUM_CMP_CNPJ.astype('str')
```

```python
df_fontes.dtypes
```

### Primeiro iremos separar em uma nova tabela os itens em que o cnpj estiver errado
### em seguida criamos uma nova apenas com os que estiverem corretos

```python
##criando nova tabela com os itens em que o campo NUM_CNPJ tiver comprimento diferente de 10
df_erroCnpj = df_fontes.loc[df_fontes['NUM_CNPJ'].str.len() != 10, ['ID_STG_FNT_ITT','NUM_CNPJ', 'NUM_CMP_CNPJ']]
```

```python
df_erroCnpj.head(10)
```

```python
df_cnpjCorreto = df_fontes.loc[df_fontes['NUM_CNPJ'].str.len() == 10, ['ID_STG_FNT_ITT','NUM_CNPJ', 'NUM_CMP_CNPJ']]
```

```python
##outra tabela, agora com os cnpj corretos
df_cnpjCorreto.head(10)
```

##
###  Em seguida verificaremos Razão social e Cnpj duplicados

```python
##total de valores Razao social duplicados
df_fontes['NOM_RAZ_SCL'].duplicated().sum()
```

```python
df_duplicados = df_fontes.loc[df_fontes['NOM_RAZ_SCL'].duplicated() == True, ['ID_STG_FNT_ITT','NUM_CNPJ', 'NUM_CMP_CNPJ', 
                                                                              'NOM_COM', 'NOM_RAZ_SCL']]
```

```python
df_duplicados.head(10)
```

```python
df_fontes['NUM_CNPJ'].duplicated().sum()
```

```python
df_duplicadosCnpj = df_fontes.loc[df_fontes['NUM_CNPJ'].duplicated() == True, ['ID_STG_FNT_ITT','NUM_CNPJ', 'NUM_CMP_CNPJ', 
                                                                              'NOM_COM', 'NOM_RAZ_SCL']]

df_duplicadosCnpj.head(10)
```

```python
df_cnpjUnico = df_fontes.loc[df_fontes['NUM_CNPJ'].duplicated() == False, ['ID_STG_FNT_ITT','NUM_CNPJ', 'NUM_CMP_CNPJ', 
                                                                              'NOM_COM', 'NOM_RAZ_SCL']]

df_cnpjUnico.head(10)
```

```python
df_razaoUnico= df_fontes.loc[df_fontes['NOM_RAZ_SCL'].duplicated() == False, ['ID_STG_FNT_ITT','NUM_CNPJ', 'NUM_CMP_CNPJ', 
                                                                              'NOM_COM', 'NOM_RAZ_SCL']]

df_razaoUnico.head(10)
```

```python
##verificando se existem ids duplicados
df_fontes['ID_STG_FNT_ITT'].duplicated().sum()
```

### Para facilitar nossa anlise iremos juntar todas as tabelas com informações erradas,
### o mesmo será feito para os dados corretos

```python
##criando dataframe vazio
df_erros = pd.concat([df_fontes['ID_STG_FNT_ITT'], df_erroCnpj['NUM_CNPJ'], df_erroCnpj['NUM_CMP_CNPJ']], axis= 1)
```

```python
df_erros.head()
```

```python
df_erros['Cnpj Invalido'] = df_erros['NUM_CNPJ'] + df_erros['NUM_CMP_CNPJ']
```

```python
df_erros.tail()
```

```python
df_erros = df_erros.drop(columns=['NUM_CMP_CNPJ'], axis=1)
df_erros.head()
```

```python
df_erros['Nome Comercial'] = df_duplicados['NOM_COM']
df_erros['Razao Social'] = df_duplicados['NOM_RAZ_SCL']
df_erros.head()
```

#### Resumo:
##### Criamos as seguintes tabelas:
##### df_fontes que seriam os dados originais
##### df_erroCnpj e df_duplicadosCnpj que são os cnpjs invalidos e duplicados respectivamente
##### df_duplicados com Razão social duplicada ou invalida
##### df_erros para resumir os principais erros
##### df_cnpjCorreto
